```python
#4
plt.figure(figsize=(8,8))
x = list(range(0,11))
y = [math.sqrt(x) for x in x]
plt.plot(x,y, label = '$sqrt(x)$', color = 'red')


x = list(range(-10,0))+ list(range(1,10))
y = [x^3 for x in x]
plt.plot(x, y, label = '$x^3$', color = 'blue')


x = list(range(-10,0))+ list(range(0,10))
y = [2^x for x in x]
plt.plot(x,y, label = '$2^x$', color = "green")



x = list(range(-10,-1))+ list(range(0,10))
y = [1/(x+1) for x in x]
plt.scatter(x,y, label = '${1}/{(x+1)}$', color = 'black')

plt.legend()



plt.show()


```

```python
import math 
import matplotlib.pyplot  as plt
import numpy as np
import pandas as pd
```

```python
#1
x = range(-50,50)
y = [x**2 - x + 2 for x in x]

plt.text(0,4,('0,0'))
plt.plot(x,y)

plt.show()
```

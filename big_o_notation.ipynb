```python
import math 
import matplotlib.pyplot as plt
```

```python

plt.figure(figsize=(16,8))
plt.title('Big O Natation')
plt.xlabel('Equations')
plt.ylabel('Operations')



x = range(1,8)
y1 = [ (0*x) + 1 for x in x]
plt.plot(x,y1,label = "$O(1)$")


y2 = [math.log(x) for x in x]
plt.plot(x,y2,label = "$O(log (n))$")

y3 = [x for x in x]
plt.plot(x,y3,label = "$O(n)$")

y4 = [x * math.log(x)  for x in x]
plt.plot(x,y4,label = "$O(n log n)$")

y5 = [x**2 for x in x]
plt.plot(x,y5,label = "$O(n^2)$")

y6 = [2**x for x in x]
plt.plot(x,y6, label = "$O(2^n)$")

y7 = [math.factorial(x) for x in x]
plt.plot(x,y7, label = "$O(n!)$")

y8 = [(x**x) for x in x]
plt.plot(x,y8, label = "$O(n^n)$")




plt.legend(bbox_to_anchor=(-.175, 0.5))
plt.xlim(1,7)
plt.ylim(1,100)



plt.show()

```

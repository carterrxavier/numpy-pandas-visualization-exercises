```python
import math 
import matplotlib.pyplot as plt
```

```python


plt.title('Big O Natation')
plt.xlabel('Equations')
plt.ylabel('Operations')



x = range(1,8)
y = [ (0*x) + 1 for x in x]
plt.plot(x,y,label = "$O(1)$")


y = [math.log(x) for x in x]
plt.plot(x,y,label = "$O(log (n))$")

y = [x for x in x]
plt.plot(x,y,label = "$O(n)$")

y = [x * math.log(x)  for x in x]
plt.plot(x,y,label = "$O(n log n)$")

y = [x**2 for x in x]
plt.plot(x,y,label = "$O(n^2)$")

y = [2**x for x in x]
plt.plot(x,y, label = "$O(2^n)$")

y = [math.factorial(x) for x in x]
plt.plot(x,y, label = "$O(n!)$")

y = [(x**x) for x in x]
plt.plot(x,y, label = "$O(n^n)$")




plt.legend(bbox_to_anchor=(-.175, 0.5))
plt.xlim(1,7)
plt.ylim(1,20)



plt.show()

```
